```python
import os
import mne
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print(__doc__)
```

```python
fifs = pd.read_json('fifs.json')
fifs
```

```python
tmin, tmax = -0.2, 1.2
picks = 'mag'
raw = mne.concatenate_raws([mne.io.read_raw_fif(fifs.loc['S02_R{:02d}'.format(j)]['icapath']) for j in range(4, 12)])
events = mne.find_events(raw, stim_channel='UPPT001')
epochs = mne.Epochs(raw, events, baseline=(tmin, 0), tmin=tmin, tmax=tmax, decim=10, picks=picks)
event_id = epochs.event_id
```

```python
n_jobs = 64
tfr_morlet = mne.time_frequency.tfr_morlet
freqs = np.logspace(*np.log10([1, 50]), num=50)
n_cycles = freqs / 2
for event in event_id:
    print(event)
    power, itc = tfr_morlet(epochs[event], freqs=freqs, n_cycles=n_cycles, use_fft=True, return_itc=True, n_jobs=n_jobs)
    power.plot_joint(title=event, timefreqs=None, baseline=(tmin, 0), tmin=tmin, tmax=tmax, mode='mean')
```
