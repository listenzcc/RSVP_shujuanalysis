```python
import os
import time
import pandas as pd
```

```python
df = pd.read_json('auto_path_table.json')
df
```

```python
new_start = False
# Save old processed_data dir
if new_start:
    if os.path.exists('processed_data'):
        os.rename('processed_data', 'processed_data_{}'.format(time.ctime()))

try:
    os.mkdir('processed_data')
except FileExistsError:
    pass
```
